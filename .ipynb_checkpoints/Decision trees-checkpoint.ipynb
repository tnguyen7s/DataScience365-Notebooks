Decision tree algorithms:
* ID3
* C4.5
* CART
* CHAID
* MARS

Solve:
* classification & regression
* work with categorical and numerical features



#### Why we choose decision tree instead of a complex and precise of a NN?
* The simplicity of this model is what makes it useful
* White box model compared to black box model (nn), easy to understand and interpret
* In-built feature selection: hierarchical structure: the root node is the most important node and the significance of the further nodes diminishes -> allows us to only focuses on the first few features
* Require little to no preprocessing and preparation (can work with categorical and numerical, don't require scaling, standardizing
* fast computation (platform likes mobile)

##### Cons
* Overfitting: learn the dataset extremely well, but also capture some random fluctuation as well => only looks for general trend\
=> Tree prunning and test on a testing set
* Unstable due to data variation: not a good option if to create a robust model not affected by a specific dataser
* TRainning through greedy algorithm: may find a locally optimal solution, but may not guarantee an optimal solution: try to find the best question to ask at each level, seperately
* Unbalanced dataset can create problem

#### Iris dataset
* Three Iris species: Iris Setosa(0), Iris Versicolor(1), Iris Virginica(2) classified based on: The length and width of petals and length and width of sepals

#### Creating a decision tree

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

### The dataset


```python
iris = load_iris()
```

```python
X, y = iris.data, iris.target
```

```python
X
# x corresponds to the width and height of sepals and petals 
```

```python
y
# corresponds to the label associated to the flower
```

```python
y.shape, X.shape
```

#### Creating a decision tree

```python
clf = DecisionTreeClassifier() # create a model
```

```python
clf = clf.fit(X, y) # train with it
```

```python
clf.predict([(6.1, 2.7, 3.9, 1.2)])
```

#### Visualizing a tree

```python
plot_tree(clf)
```

```python
plt.figure(figsize=(15,12))
plot_tree(clf, filled=True)
plt.show()
```

 * Leaves: final outcome
 * other nodes conain question
 * the arrows are answers to the question

X[2] <= 2.45
* X[2] the third feature in our input
* Question whether the value corresponding to the third feature <= 2.45


Samples=150: there are 150 samples reach the first node (obviously)

Value = [50,50,50]: number ofsample in each of the class

if X[2]<=2.45, go to the orange leave, which classes it belongs?
* by looking at value=[50,0,0], means that the 50 data points of the first class has been classified\
=> We can say that if the third feature of the flower less than 2.45, it will be classified as Iris Setosa

```python
plt.figure(figsize=(15,12))
plot_tree(clf, 
          filled=True, 
          feature_names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'], 
         class_names=['Setosa', 'Versicolour', 'Virginica'])
plt.show()
```

 #### How do we define what is the best node? There are metrics to quantify how good or bad a certain node is?
 * Gini impurity
 * Gin Information

Gini impurity: used by the CART algorithm: how much changes in the node (its accuracy to split the data with an answer yes or no) when we shuffled the data ->calculate the misclassification rate

* The algorithm will try and find a split with the lowest combined Gini for both child nodes
* sum (Pi*(Pi-1))

gini=0.168=49/54*(1-49/54)+5/54*(1-5/54)

Information gain: used by ID3, C4.5:
* sum(pi*log2_pi): pi is the percentage that each class presents in the child node that comes from a split


Gini impurity and Gain Informatio both good but Gain Information may require additional computation cost

#### Dealing with overfitting: prunning
* Algorithm aims to achieve a pure split: a node of single class only
* Avoid exccessive detail
=> Prunnning: remove parts that are not neccessary for final classification, reduce the complexity, import the predict accuracy

Minimal Cost complexity prunning: identify the subtree with the smallest contribution and cut from the tree
* Hyperparameter alpha: how strong the prunning: 0(no prunning), >1(severe prunning) may be 0.1, 0.01, 0.001

## Random Forest

* Very good accuracy for a non-neural network model
* Decision tree looses the random forest

A forest is a collection of decision trees\
Ensembling modeling: uses multiple learning algorithms to obtain better predictive performance \
Rely on the wisdom of the crowd \
Uninterpretable and not transparent: black model \
Each decision tree in the random forest produces a different decision => uses the majority vote

#### Bootstrapping

Aims to create many different datasets of the same size follows the same distribution of the original dataset but have minor changes => each dataset obtains different decision tree\
Uniformly sampling from the original dataset with replacement\
When performing, it is expected the new dataset contains 1-1/e=63% unique samples

##### Bootstrapping Aggregating/Bagging

100 decision trees trainned from 100 samples. What do we do with them?  
* Combine output via majority voting
* random forest don't consider all the features at once

## Census data and income
Input features: census data \
Output: outcome \
Train set 2/3 + Test set 1/3 (49000 data points)

```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("datasets/Census_income_train.csv")
```

```python
train_data.head()
```

```python
train_data.shape
```

```python
train_data.isnull().sum()
```

##### Remove rows with unknown values

```python
clean_train_data  = train_data[~train_data["Workclass"].str.contains("\?")]
```

```python
clean_train_data = clean_train_data[~clean_train_data["Occupation"].str.contains("\?")]
```

```python
clean_train_data = clean_train_data[~clean_train_data["Native-country"].str.contains("\?")]
```

```python
clean_train_data.reset_index(drop=True)
```

#### Create dummies

```python
train_dummies = pd.get_dummies(clean_train_data, drop_first=False)
```

```python
train_dummies.head()
```

```python
del train_dummies["Income_ <=50K"]
```

```python
inputs = train_dummies.iloc[:, :-1]
target = train_dummies.iloc[:, -1]
```

```python
inputs.head()
```

```python
target.head()
```

```python
test_data = pd.read_csv("datasets/Census_income_test.csv")
clean_test_data  = test_data[~test_data["Workclass"].str.contains("\?")]
clean_test_data = clean_test_data[~clean_test_data["Occupation"].str.contains("\?")]
clean_test_data = clean_test_data[~clean_test_data["Native-country"].str.contains("\?")]
clean_test_data.reset_index(drop=True)
test_dummies = pd.get_dummies(clean_test_data, drop_first=False)
del test_dummies["Income_ <=50K."]
test_inputs = test_dummies.iloc[:, :-1]
test_target = test_dummies.iloc[:, -1]
```

### Creating a decision tree

```python
clf = tree.DecisionTreeClassifier(random_state=365, ccp_alpha=0.001)
```

```python
clf = clf.fit(inputs, target)
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(clf, filled=True)
plt.show(clf)
```

```python
### Testing the model
test_pred = clf.predict(test_inputs)
```

```python
print(classification_report(test_target, test_pred))
```

### Creating a random forest: random forest is naturally resistant to overfitting

```python
clf = RandomForestClassifier(random_state=365, n_estimators=150, ccp_alpha=0.0001)
clf.fit(inputs, target)
```

```python
 test_pred = clf.predict(test_inputs)
```

```python
print(classification_report(test_target, test_pred))
```
