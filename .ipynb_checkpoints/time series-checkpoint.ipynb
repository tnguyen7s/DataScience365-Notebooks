#### Stationary

Weak form stationary - covariance stationary: Taking consecutive data of the same interval should give the same covariance. Statisfy 3 assumptions:
* Constant mean
* Constant variance
* Cov(xn, xn+k) = Cov(xm, xm+k) such as Cov(x1, x4) = Cov(x3, x6)

An example of weak form stationary is white noise 
* mean=0
* constant standard deviation
* cov(xn, xn+k)=0 since cov=corr*std1*std2 and corr=0 => covariance is not the function of time


Strict stationary: identical sample size should have identical distribution
* dist(xt, xt+k) = dist(xt+tau, xt+tau+k)

##### Dickey Fuller test

* H0: assumes non stationary T, the 1-lag auto correlation coefficient is lower than 1
* H1: 1-lag auto correlation coefficient = 1
* if the test statistics < critical value => reject the H0

```python
import statsmodels.tsa.stattools as sts
```

```python
sts.adfuller(df.market_value)
```

* t-statistics > each of the critical value => we do not find sufficient evidence of stationary
* p_value=0.4 => 40% likelihood of observing such result if H1 is true
* 18 is the number of lags used in the regression. there is some autocorrelation going back 18 periods (correlation between values 18 periods apart)

```python
sts.adfuller(df.wn) 
```

* expect to get low p value since wn is stationary 
* since wn is no auto correlation => expect number of lags=0

```python
sts.adfuller(df.rw)
```

* 60% chance the data comes from a non-stationary process
* values go up and down but the starting position is different
* The covariances of the two same size intervals rarely identical

#### Correlations between past and present values
Correlation: Measures the similarity in the change of values of two series\
Autocorrelation: the correlation  between a sequence and the laged version of itself

##### Autocorrelation function (ACF):
Computes the autocorrelation value for many lags we are interested simultanesouly
* p(xt, xt-1), p(xt, xt-2), p(xt, xt-3)
* find the correlation between the data and many lagged versions of itself.

```python
import statsmodels.graphics.tsaplots as sgt
sgt.plot_acf(df.market_value, lags=40, zero=False) # autocorrelation of itself with 40 lagged versions
plt.title('ACF S&P', size=24)
plt.show()
```

* the line is the auto correlation one time period ago, two time periods ago, and so on -> suggests the existence of auto correlation
* the blue area is the significance: the greater the distance in time, the more unlikely that the auto correlation persist.
-> Make sure that the correlation coefficient in higher lags is sufficiently greater to be significantly different from 0
-> As the graph suggests, all the lines are greater than the blue region, thus the correlation coefficients are significant
-> indicator of time dependence
-> the prices even a month ago can still serve as decent estimators

```python
sgt.plot_acf(df.wn, lags=40, zero=False) # autocorrelation of itself with 40 lagged versions
plt.title('ACF WN', size=24)
plt.ylim(-0.2,0.2)
plt.show()
```

* All the line fall within the blue area
* The coefficient are insignificant -> no autocorrelation for any lag

```python
sgt.plot_acf(df.rw, lags=40, zero=False)
plt.title('ACF RW', size=24)
plt.show()
```

#### PACF

The correlation coefficient measures:
1. the similarity between itself the its lagged version
2. the second effect, for example, the third lag coefficient captures  "direct and indirect ways" in which the lag series affect the current

To determine the direct relationship between the time series and its lagged version, we need to compute the partial auto correlation

```python
sgt.plot_pacf(df_comp.market_value, lags=40, zero=False, method="ols")
plt.title("PACF S&P", size=40)
plt.show()
```

* only the first version (one lag behind) has the coefficient that is significantly different from 0
* for instance, measures the effect of values in period T-2 have on the values in period T directly
*  The ACF measures the accumulated effects past lags have on the current value, while the PACF measures the direct effects.

```python
sgt.plot_pacf(df.wn, lags=40, zero=False, method="ols")
plt.title("PACF WN", size=40)
plt.show()
```

## Autoregressive model:  it uses a lagged version of itself (auto) to conduct the regression.

Relies on past period values to predict current value: 

A linear model where current period values are a sum of past outcomes multiply by a numeric factor\
[x in t] = C (constant) + phi (coefficient) * [x in (t-1)] + epsilonT

* xt-1: the value of x during the previous period
* phi: part of the previous period which remains in the future (between -1 and 1)
* epsilonT: Residual - the difference between the prediction value and the actual value\
=> The value at a given period equals some portions of the previous period + some constant benchmark + unpredictable shock

A model using 2 lags: x(t) = C + phi1 * x(t-1) + phi2 * x(t-2) + epsilonT


More lags -> More complicated -> More coefficients -> More likely not significant

If the coefficient is not significantly different from 0, they would have no  effect on the predicted values

Since this happens to be the case, we need to use LOG-LIKELIHOOD RATIO (LLT) test to determine whether the two models are significantly different


#### LLR test

```python
from scipy.stats import chi2
```

```python

def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    
    LR = (2*(L2-L1)) # compute the test statistics
    p= chi2.sf(LR, DF).round(3)
    return p
```

#### Comparing  Higher Lag AR models

```python
LLR_test(model_ar_2, model_ar_3)
```

* The difference in loglikehood is significant -> we should opt for a more complicated model

```python
model_ar_6 = ARIMA(df.market_values, order=(6,0,0))
results_ar_6 = model_ar_6.fit()
model_ar_7 = ARIMA(df.market_values, order=(7,0,0))
results_ar_7 = model_ar_7.fit()
print(results_ar_6.summary())
print(results_ar_7.summary())
print('LLR Test: ', LLR_test(model_ar_6, model_ar_7))
```

```python
model_ar_7 = ARIMA(df.market_values, order=(7,0,0))
results_ar_7 = model_ar_7.fit()
model_ar_8 = ARIMA(df.market_values, order=(8,0,0))
results_ar_8 = model_ar_8.fit()
print(results_ar_7.summary())
print(results_ar_8.summary())
print('LLR Test: ', LLR_test(model_ar_7, model_ar_8))
```

# AM model

There exist a type of self correcting model which takes into past account residuals

They do exceptionally well with random walk dataset because they always adjust from the errors of previous periods

rt = c + phi1 * epsilon(t-1) + epsilon(t)

* rt is the value of r in current period
* a numeric coefficient for the value associated  with the 1st lag
* epsilon(t) the residual for the current period
* epsilon(t-1) the residul for the previous perio

Compare between autoregressive and moving average:
* y = c + phi1 * y(previous) + epsilon rely on the past values
* y = c + phi1 * epsilon(previous) + epsilon rely on the past residuals
The coefficient should be less than 1

Difference between autoregression and average moving models:
* Autoregressive: relies on PACF
* Average Moving: relies on ACF since the MA model aren't reliant on past periods' values, DONT'T need to determine which lag has significant DIRECT return on the current value

ACF gives information about how many lags that the model should use: about 8 lags

### ARMA (Auto regressive - Moving Average model)

* AR: fail to adjust quickly for unexpected shock
* MA: account only for residual\
=> Model takes into account the past values and past errors
* take into account the actual result
* but also how far our prediction


r_t = c + phi1 * r_(t-1) + theta1 * epsilon_(t-1) + epsilon_(t)\
* What part of the value past period is relevant in explainning the current one
* What part of the error past period is relevant in explainning the current one

ARMA(P, Q)
* Take values upto P periods ago
* Take values upto Q periods ago

## ARIMA model for non stationary data (Auto Resgressive Integrated Moving Average)

The integrated model accounts for the number of non-seasonal differences we need to examine to establish stationary

It estimates the jump from one value to the next period's value instead of the actual values

Ensure the stationarity no matter the underlying data

Transform from price to price return is an integration

ARIMA(p,d,q)
* p -> AR lags
* q -> MA lags
* d -> The number of times we need to integrate the time series to ensure stationarity

For the AR component, use the price differences instead of prices
* We are integrating d many times to construct new time series and then fit that to ARMA(p,q) model

delta_p_in_t = c + phi1 * delta_p_in_t-1 + beta1 * epsilon_in_t-1 + epsilon_in_t\
* phi1: what part of the previous period value is relevant in explainning the current value
* beta1: what part of the previous period error term is relevant in explainning the current value
* delta_p_in_t is the difference in prices of current period and its preceding period

ARIMA model is a ARMA(p,q) model with newly generated stationary data, and the number of integration applied on the time series can be 2,3,4...

ARIMA(p,2,q) for P = ARIMA(p,1,q) for Q where Q is the difference between the previous value and the current value

Examine the ACF for the residuals and gets a better feel about which lags to use

Lose several observations when dealing with integrated values. No previous value for first observation.\
If we integrate twice, we lose the first two observations

## ARIMAX model

Model that takes into account more than past values and past errors

delta_p_in_t = c + beta * X + phi1 * delta_p_in_t-1 + theta1 * epsilon_in_t-1 + epsilon_in_t

* Theta is a value fitted based on the data and model selection
* X: can be any variable, such as a time varying measurement (such as inflation rate), categorical data, boolean value, ..., a combinations of factors that affects the model. called exogeneous variables

```python
model_ar_1_i_1_ma_1_Xspx = ARIMA(df.market_values, order=(1,1,1), exog=raw_csv_data.spx)
result_ar_1_i_1_ma_1_Xspx = model_ar_1_i_1_ma_1_Xspx.fit()
```
